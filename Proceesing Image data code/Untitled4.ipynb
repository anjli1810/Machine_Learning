Resizing the image using Pillow Library

```python
from PIL import Image
```

```python
img = Image.open('/content/dog.jpg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('dog_image_resize.jpg') #saving the new resized image
```

```python
img_res=mpimg.imread('/content/dog_image_resize.jpg')
#Displaying the image from numpy array
img_plot=plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

Converting RGB images sto grayscale image using OpenCV

```python
#importing openCV library
import cv2
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
print(grayscale_image.shape) # no 3rd value
```

```python
# cv2.imshow() -> collab does not allow this
from google.colab.patches import cv2_imshow
```

```python
#displaying the gray scale image
cv2_imshow(grayscale_image)
```

```python
#saving the grayscale image
cv2.imwrite('dog_grayscale_image.jpg',grayscale_image)
```
